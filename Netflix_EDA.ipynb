**The goal is to clean and preprocess the Netflix dataset to make it ready for meaningful analysis. The dataset provides detailed information about shows and movies available on Netflix, including metadata such as directors, cast members, countries of production, and genres. By addressing missing values, standardizing formats, and extracting useful insights, we aim to uncover trends and patterns in the Netflix catalog.**

# Key Features

show_id: A unique identifier for each Netflix title.

type: Specifies whether the title is a "Movie" or a "TV Show".

title: The name of the show or movie.

director: The name of the director (if available).

cast: A list of cast members involved in the production.

country: The country where the show or movie was produced.

date_added: The date the title was added to Netflix's catalog.

release_year: The year the title was originally released.
rating: The age rating or maturity level (e.g., PG, TV-MA, etc.).

duration: Indicates either the duration of a movie in minutes or the number of seasons for a TV show.

listed_in: Categories or genres associated with the content (e.g., Drama, Comedy).

description: A brief summary or synopsis of the show or movie.

Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
get_ipython().run_line_magic('matplotlib', 'inline')
```

Data Extraction

```python
netflix_data = pd.read_csv("/content/netflix_titles.csv")
netflix_data.head()
```

```python
netflix_data.info()
```

Handle missing values

```python
netflix_data['director'].fillna('Unknown', inplace=True)
netflix_data['cast'].fillna('Unknown', inplace=True)
netflix_data['country'].fillna('Unknown', inplace=True)
netflix_data['date_added'].fillna('Unknown', inplace=True)
```

```python
netflix_data.head()
```

Remove duplicates

```python
netflix_data.drop_duplicates(inplace=True)
netflix_data['date_added'] = pd.to_datetime(netflix_data['date_added'], errors="coerce")
netflix_data['year_added'] = netflix_data['date_added'].dt.year
netflix_data['month_added'] = netflix_data['date_added'].dt.month
```

```python
netflix_data['date_added']
```

Filling NaN values in newly created columns

```python
netflix_data['year_added'].fillna(0, inplace=True)
netflix_data['month_added'].fillna(0, inplace=True)
```

Transform duration for analysis (handle movies vs TV shows)

```python
def transform_duration(duration):
    if isinstance(duration, str):
        if 'Season' in duration:
            return int(duration.split()[0])
        elif 'min' in duration:
            return int(duration.split()[0])
    return np.nan
```

```python
netflix_data['duration_numeric'] = netflix_data['duration'].apply(transform_duration)
```

```python
netflix_data.head()
```

EXPLORATORY DATA ANALYSIS
YEARLY CONTENT PRODUCTION TREND

```python
plt.figure(figsize=(12,6))
netflix_data['release_year'].value_counts().sort_index().plot(kind='line',color="blue")
plt.title('Yearly Content Production on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

DISTRIBUTION OF CONTENT TYPES(MOVIE VS TV SHOWS)

```python
plt.figure(figsize=(8,5))
sns.countplot(data=netflix_data, x='type', palette='coolwarm')
plt.title('Distribution of Content Type (Movies vs TV Shows)')
plt.show()
```

TOP 10 COUNTRIES BY CONTENT PRODUCTION

```python
plt.figure(figsize=(10,6))
top_countries = netflix_data['country'].value_counts().head(10)
top_countries.plot(kind='bar', color='green')
plt.title('Top 10 Countries by Content Production')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.show()
```

TOP 10 MOST COMMON GENRES

```python
plt.figure(figsize=(10,6))
netflix_data['listed_in'].value_counts().head(10).plot(kind='barh', color='purple')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.show()
```

DISTRIBUTION OF RATINGS

```python
plt.figure(figsize=(10,6))
netflix_data['rating'].value_counts().plot(kind='bar', color='orange')
plt.title('Distribution of Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.show()
```

ADVANCED ANALYSIS
CORRELATION BETWEEN VARIABLES(NUMERIC DURATION VS RELEASE YEAR)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filtered_data = netflix_data.dropna(subset=['release_year', 'duration_numeric'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=filtered_data, x='release_year', y='duration_numeric', hue='type', palette='cool')
plt.title('Correlation Between Release Year and Duration')
plt.xlabel('Release Year')
plt.ylabel('Duration (Minutes or Seasons)')
plt.grid(True)
plt.show()
```

DISTRIBUTION OF CONTENT BY RELEASE YEAR AND COUNTRY

```python
plt.figure(figsize=(12,8))
sns.heatmap(pd.crosstab(netflix_data['release_year'], netflix_data['country'].fillna('Unknown')), cmap='viridis')
plt.title('Heatmap of Content by Release Year and Country')
plt.show()
```
